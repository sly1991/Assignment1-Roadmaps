#Step1

```python
import urllib2
url='http://www.cricbuzz.com/cricket-schedule/upcoming-series'
page=urllib2.urlopen(url)
from bs4 import BeautifulSoup as BS
soup=BS(page)
```

```python
international_element=soup.find_all(attrs={"ng-show":"((filtered_category == 0 || filtered_category == 9))"})
#soup.select('div[ng-show^="((filtered_category == 0 || filtered_category == 9))"]')
#series=soup.find_all('div',class_="cb-col-100 cb-col",ng-show_="((filtered_category == 0 || filtered_category == 9))")    
```

```python
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
for i in international_element:
    date=i.find('div',class_="cb-lv-gray-strip text-bold")
    A.append(date.getText())
    
    series=i.find('div',class_="cb-col-33 cb-col cb-mtchs-dy text-bold")
    B.append(series.getText())
    
    match=i.find('a')
    C.append(match.getText())
    
    location=i.find('div',class_="cb-font-12 text-gray cb-ovr-flo")
    D.append(location.getText())
    
    time=i.find('div',class_="cb-font-12 text-gray")
    E.append(time.getText())
    
```

```python
import pandas as pd
df=pd.DataFrame(A,columns=['Date'])
df['Series']=B
df['Match']=C
df['Location']=D
df['Time']=E
df
```

#Step2
Sorting the dataframe based on location and no.of matches in that location

```python
count_dict=dict()
for i in newest_df.Location:
    if i in count_dict:
        count_dict[i]=count_dict[i]+1
    else:
        count_dict[i]=1
```

```python
keys=count_dict.keys()
keys
values=[]
for i in keys:
    value=count_dict[i]
    values.append(value)
    
```

```python
pop_df=pd.DataFrame(keys,columns=['Location'])
pop_df['number of matches']=values
```

```python
import json
import urllib
serviceurl = 'https://maps.googleapis.com/maps/api/geocode/json?'

lat_co=[]
long_co=[]

for i in range(len(pop_df.Location)):
    loc=pop_df.Location[i]
    url = serviceurl + urllib.urlencode({'address':loc})
    key='&key='
    url_key=url+key
    uh = urllib.urlopen(url_key)
    data = uh.read()
    js = json.loads(str(data))
    js = json.loads(str(data))
    lat = js["results"][0]["geometry"]["location"]["lat"]
    lng = js["results"][0]["geometry"]["location"]["lng"]
    lat_co.append(lat)
    long_co.append(lng)
pop_df['latitude']=lat_co
pop_df['longitude']=long_co
pop_df
```

```python
sorted_df=pop_df.sort(['number of matches'],ascending=[True])
sorted_df
```

#Final Data Frame (after sorting in increasing no. of matches) 

```python
df=pd.DataFrame(list(pop_df.Location),columns=['Location'])
df['no. of matches']=list(pop_df['number of matches'])
df['lat']=list(pop_df.latitude)
df['long']=list(pop_df.longitude)
df
```

#Step4 
Bubble map plotting using plotly api

```python
import plotly.plotly as py
import pandas as pd

lst=[]
n=len(df.Location)
for i in range(n):
    lst.append(str(df['no. of matches'][i])+'match/matches')

df['text'] = lst
limits = [(0,16),(17,21),(22,26),(27,32),(40,41)]
colors = ["rgb(0,116,217)","rgb(255,65,54)","rgb(133,20,75)","rgb(255,133,27)","lightgrey"]
locations = []
scale = 50

for i in range(len(limits)):
    lim = limits[i]
    df_sub = df[lim[0]:lim[1]]
    loc = dict(
        type = 'scattergeo',
        lon = df_sub['long'],
        lat = df_sub['lat'],
        text = df_sub['text'],
        marker = dict(
            size = df_sub['no. of matches']*scale,
            color = colors[i],
            line = dict(width=0.5, color='rgb(40,40,40)'),
            sizemode = 'area'
        ),
        name = '{0} - {1}'.format(lim[0],lim[1]) )
    locations.append(loc)

layout = dict(
        title = 'Number of international matches in different locations from July 2016 to January 2017 ',
        showlegend = True,
        geo = dict(
            showland = True,
            landcolor = 'rgb(217, 217, 217)',
            subunitwidth=1,
            countrywidth=1,
            subunitcolor="rgb(255, 255, 255)",
            countrycolor="rgb(255, 255, 255)"
        ),
    )

fig = dict( data=locations, layout=layout )
url = py.plot( fig, validate=False, filename='d3-bubble-map-No.of mathces' )
```
